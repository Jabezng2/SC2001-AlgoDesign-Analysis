<h3>Import libraries</h3>

```python
import random
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

<h3>Function to generate Adjacency Matrix</h3>

```python
def generateAdjMatrix(num_of_vertices):
    adj_matrix = []
    for i in range(num_of_vertices):
        adj_lst = []
        if i == 0:
            adj_lst = [0] + [random.randint(1,10) for x in range(num_of_vertices-1)]
            adj_matrix.append(adj_lst)
        else:
            j = 0
            while j < i:
                adj_lst.append(adj_matrix[j][i])
                j += 1
            adj_lst.append(0)
            adj_lst += [random.randint(2,10) for x in range(num_of_vertices-1 - i)]
            adj_matrix.append(adj_lst)
    return adj_matrix
```

<h3>Graph with 3 vertices</h3>

```python
num_of_vertices = 3

adj_matrix = generateAdjMatrix(num_of_vertices)
G = nx.Graph()
for i in range(len(adj_matrix)):
    for j in range(i, len(adj_matrix[i])):
        if adj_matrix[i][j] != 0:
            G.add_edge(i, j, weight=adj_matrix[i][j])
            
plt.figure(figsize=(10,10))
pos = nx.spring_layout(G, seed=num_of_vertices)
nx.draw(G, pos, with_labels = True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, "weight"))
```

```python
print("The Adjacency Matrix for Graph 1 is:")
for lst in adj_matrix:
    print(lst)
```
