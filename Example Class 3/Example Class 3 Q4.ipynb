<b>Import essential libraries</b>

```python
import pandas as pd
from IPython.display import display
```

<b>Knapsack bottom-up DP algorithm </b>

```python
def knapsack_dp(obj_wgts_lst, obj_profits_lst, capacity):
    #get number of objects
    num_of_obj = len(obj_wgts_lst)
    
    #initialising 2D array tracking subsets' profits
    profits = [[0 for obj_num in range(num_of_obj+1)] for size in range(capacity+1)]
    
    for size in range(1, capacity+1): #for each subsize of capacity
        for obj_num in range(1, num_of_obj+1): #for each object
            
            #safe choice : not taking this object means knapsack has not exceeded capacity
            profits[size][obj_num] = profits[size][obj_num-1]
            
            #get current object's weight and profit
            cur_obj_wgt = obj_wgts_lst[obj_num-1]
            cur_obj_profit = obj_profits_lst[obj_num-1]
            
            #check for two conditions:
            #(1) the weight of current object is able to fit into current subsize
            # (2) net profit achieved by taking the current object and possibly not taking some other objects
            if cur_obj_wgt <= size:
                if profits[size][obj_num] < profits[size-cur_obj_wgt][obj_num] + cur_obj_profit:
                    #conditions satisfied, for current subsize, taking current object will provide more profits than not taking
                    profits[size][obj_num] = profits[size-cur_obj_wgt][obj_num] + cur_obj_profit
            
            #visualise profits list to show running order
            df = pd.DataFrame(profits)
            print("For sub-capacity", size, "and object number", obj_num)
            display(df)
            print("\n")
    
    return profits[capacity][num_of_obj]
```

<b> Using arrays in part A </b>

```python
w = [4, 6, 8]
p = [7, 6, 9]

max_profit = knapsack_dp(w, p, 14)
print("######### Therefore the max profit is", max_profit, "#########")
```

<b> Using arrays in part B </b>

```python
w = [5, 6, 8]
p = [7, 6, 9]

max_profit = knapsack_dp(w, p, 14)
print("######### Therefore the max profit is", max_profit, "#########")
```

<b> Proposed modification to show objects taken </b>

```python
def knapsack_dp_modified(obj_wgts_lst, obj_profits_lst, capacity):
    #get number of objects
    num_of_obj = len(obj_wgts_lst)
    
    #initialising 2D array tracking subsets' profits and objects taken
    profits = [[0 for obj_num in range(num_of_obj+1)] for size in range(capacity+1)]
    obj_takens = [[[] for obj_num in range(num_of_obj+1)] for size in range(capacity+1)]
    
    for size in range(1, capacity+1): #for each subsize of capacity
        for obj_num in range(1, num_of_obj+1): #for each object
            
            #safe choice : not taking this object means knapsack has not exceeded capacity
            profits[size][obj_num] = profits[size][obj_num-1]
            obj_takens[size][obj_num] = obj_takens[size][obj_num-1]
            
            #get current object's weight and profit
            cur_obj_wgt = obj_wgts_lst[obj_num-1]
            cur_obj_profit = obj_profits_lst[obj_num-1]
            
            #check for two conditions:
            #(1) the weight of current object is able to fit into current subsize
            # (2) net profit achieved by taking the current object and possibly not taking some other objects
            if cur_obj_wgt <= size:
                if profits[size][obj_num] < profits[size-cur_obj_wgt][obj_num] + cur_obj_profit:
                    #conditions satisfied, for current subsize, taking current object will provide more profits than not taking
                    profits[size][obj_num] = profits[size-cur_obj_wgt][obj_num] + cur_obj_profit
                    obj_takens[size][obj_num] = obj_takens[size-cur_obj_wgt][obj_num] + [obj_num]
    
    #display tables
    display(pd.DataFrame(profits))
    display(pd.DataFrame(obj_takens))
    
    return profits[capacity][num_of_obj]
```

```python
w = [4, 6, 8]
p = [7, 6, 9]

max_profit = knapsack_dp_modified(w, p, 14)
print("######### Therefore the max profit is", max_profit, "#########")
```

```python
w = [5, 6, 8]
p = [7, 6, 9]

max_profit = knapsack_dp_modified(w, p, 14)
print("######### Therefore the max profit is", max_profit, "#########")
```
