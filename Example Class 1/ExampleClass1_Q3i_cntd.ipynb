```python
#libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
sb.set()
```

```python
csv_data = pd.read_csv("keycomparisons.csv")
csv_data.head()
```

```python
datainputs = pd.DataFrame(csv_data['Data Inputs'])
datainputs.head()
```

```python
theoreticalBest = []
theoreticalAvg = []
theoreticalWorst = []
data = 1000
for i in range(5):
    theoreticalBest.append((int)(data*math.log2(data/7)+data))
    theoreticalAvg.append((int)(data*math.log2(data/7)+data*7))
    theoreticalWorst.append((int)(data*math.log2(data/7)+data*7))
    data*=10
csv_data.insert(loc=4, column='TheoreticalBest', value=theoreticalBest)
csv_data.insert(loc=5, column='TheoreticalAvg', value=theoreticalAvg)
csv_data.insert(loc=6, column='TheoreticalWorst', value=theoreticalWorst)
csv_data.head()
```

```python
csv_data['Data Inputs'] = csv_data['Data Inputs'].apply(str)

f = plt.figure(figsize = (16,16))
ax = f.add_subplot()
ax.plot(csv_data['Data Inputs'], csv_data['Best'], color = "blue")
ax.plot(csv_data['Data Inputs'], csv_data['Average'], color = "orange")
ax.plot(csv_data['Data Inputs'], csv_data['Worst'], color = "red")
ax.legend(['Best', 'Average', 'Worst'])

plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))


ax.set_xlabel("Data Inputs")
ax.set_ylabel("Key Comparisons")
ax.set_title("HybridSort Key Comparisons Against Input Size")
```
